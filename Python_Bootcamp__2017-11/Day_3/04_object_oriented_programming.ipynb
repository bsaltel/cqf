## Doing More Backtests

```python
bt = mlb.ScikitBacktester()
```

```python
symbols = ['AAPL.O', 'MSFT.O', 'INTC.O']
for sym in symbols:
    bt.run_strategy(sym, '2010-1-5', '2017-10-31', 'LogisticRegression', lags=10)
    bt.plot_results()
```

```python
symbols = ['AAPL.O', 'MSFT.O', 'INTC.O', 'GLD']
for sym in symbols:
    bt.run_strategy(sym, '2013-1-5', '2017-10-31', 'SVC', lags=10)
    bt.plot_results()
```

```python
symbols = ['XAU=', 'GDX', 'GLD']
for sym in symbols:
    bt.run_strategy(sym, '2016-1-5', '2017-10-31', 'NaiveBayes', lags=5)
    bt.plot_results()
```
